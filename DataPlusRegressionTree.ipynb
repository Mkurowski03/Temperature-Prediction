## Regression Tree for spring as 3rd, 4th and 5th month 

```python
X = wth_345[["Summary", "Precip Type", "Humidity", "Visibility (km)", "Pressure (millibars)", "hour", "Month"]]
y = wth_345["Temperature (C)"]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
wth_345.drop(["Formatted Date", "Wind Speed (km/h)", "Wind Bearing (degrees)"], axis=1, inplace=True) 
```

```python
wth_345_encoded = pd.get_dummies(wth_345, columns=["Summary", "Precip Type", "hour", "Month"]) 
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
hyper_345 = pd.DataFrame(columns=["depth", "mse", "score"]) #
for i in range(1, 100, 5):
    X_train, X_test, y_train, y_test = train_test_split(wth_345_encoded.drop("Temperature (C)", axis=1), wth_345_encoded["Temperature (C)"], test_size=0.2)
    regressor = DecisionTreeRegressor(max_depth=i)
    regressor.fit(X_train, y_train)
    y_pred = regressor.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    score = regressor.score(X_test, y_test)
    new_row = [i, mse, score]
    hyper_345.loc[len(hyper_345)] = new_row
ax = hyper_345.plot(x="depth", y="mse", legend=False)
ax2 = ax.twinx()
hyper_345.plot(x="depth", y="score", ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.show()
```

```python
hyper_345.sort_values(by=["mse", "score"], ascending=True)[:10] #depth 11, mse 13.68
```

## Regression Tree for spring as 2nd, 3rd and 4th month 

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


X = wth_234[["Summary", "Precip Type", "Humidity", "Visibility (km)", "Pressure (millibars)", "hour", "Month"]]
y = wth_234["Temperature (C)"]
wth_234.drop(["Formatted Date", "Wind Speed (km/h)", "Wind Bearing (degrees)"], axis=1, inplace=True)
wth_234_encoded = pd.get_dummies(wth_234, columns=["Summary", "Precip Type", "hour", "Month"])
hyper_234 = pd.DataFrame(columns=["depth", "mse"])
hyper_234 = pd.DataFrame(columns=["depth", "mse", "score"])

summary_df = pd.DataFrame()
for j in range (1, 30):
    for i in range(1, 30):
        X_train, X_test, y_train, y_test = train_test_split(wth_234_encoded.drop("Temperature (C)", axis=1), wth_234_encoded["Temperature (C)"], test_size=0.2)
        regressor = DecisionTreeRegressor(max_depth=j)
        regressor.fit(X_train, y_train)
        y_pred = regressor.predict(X_test)
        mse = mean_squared_error(y_test, y_pred)
        score = regressor.score(X_test, y_test)
        new_row = [i, mse, score]
        hyper_234.loc[len(hyper_234)] = new_row
        summary_df = pd.concat([summary_df, pd.DataFrame({"mse": [mse], "max_depth" : [j]})])
grouped_hyper_234 = hyper_234.groupby("depth").mean()
grouped_hyper_234.set_index("mse")
ax = grouped_hyper_234.plot(y="mse", legend=False)
ax2 = ax.twinx()
grouped_hyper_234.plot(y="score", ax=ax2, legend=False, color="r")
ax.figure.legend()
plt.show()
```

```python
grouped_hyper_234.sort_values(by=["mse", "score"], ascending=True)[:10] #depth 9 mse 20
```
